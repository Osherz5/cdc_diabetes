## Project Topic

This supervised machine learning project aims to analyze the CDC diabetes dataset to predict the likelihood of diabetes in individuals based on various health and demographic factors.
By leveraging features such as age, BMI, blood pressure, and cholesterol levels, we employ ensemble based algorithms to classify participants as diabetic or non-diabetic.
The project includes data preprocessing, feature selection, model training, and evaluation using various metrics.
Ultimately, the goal is to develop a reliable predictive model that can assist healthcare professionals in early identification and intervention for diabetes risk, contributing to better management and prevention strategies.


<a id="org12f8614"></a>

# Data

The Diabetes Health Indicators Dataset includes 253680 records with 21 predictors consisting of healthcare statistics and lifestyle survey data, featuring attributes such as demographics, lab test results, and survey responses.
The target variable indicates whether a patient is diabetic/pre-diabetic, or healthy.

# Model selection
Given what we've seen so far, we want to use a model that can handle imbalanced classes and is robust enough for handling features collinearity.
For this purpose I'd like to try Random Forest and Gradient Boosting (LightGBM).



```python
X = df.drop(['Diabetes_binary'],axis=1)

y = df['Diabetes_binary']
```

```python
# Re-balance the dataset by undersampling

n_samples = y.sum()

y = pd.concat([(y[y==0]).sample(n_samples),(y[y==1])])

X = X.loc[y.index]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42) # Split into train and test+eval
X_test, X_eval, y_test, y_eval = train_test_split(X_test, y_test, test_size=0.5, random_state=42) # Split test and eval

# Datasets are 0.6 training, 0.2 eval (cv), 0.2 test
```

```python
print(f"Training dataset shape: {X_train.shape}")
print(f"Validation dataset shape: {X_eval.shape}")
print(f"Testing dataset shape: {X_test.shape}")
```

## Random Forest

```python
rf_clf = RandomForestClassifier()

grid_cv = GridSearchCV(rf_clf, param_grid={
                                        "n_estimators": [10,50,100,150], # Anything above takes a really long time to compute and doesn't result in a significant increase of cv score
                                        },
                                    n_jobs=-1,
                                    verbose=True,
                                    cv=5
                           )
```

```python
grid_res = grid_cv.fit(pd.concat([X_train, X_eval]),pd.concat([y_train, y_eval])) # We can't tell sklearn GridSearchCV which cv dataset to use, so we will merge these instead, to have a similar proportion as the next model
```

```python

plt.plot([params['n_estimators'] for params in grid_res.cv_results_['params']],
        [score for score in grid_res.cv_results_['mean_test_score']])
plt.xlabel("n_estimators")
plt.ylabel("cv_score")
plt.title("Random-Forest Classifier CV")

```

```python
# Predict on test-set
best_rf = grid_cv.best_estimator_
y_pred = best_rf.predict(X_test)
```

```python
# Evaluation
print(f"Confusion matrix: \n {confusion_matrix(y_test, y_pred>0.5)}")

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve");


best_rf_acc = accuracy_score(y_test, y_pred > 0.5)
best_rf_f1 = f1_score(y_test, y_pred > 0.5)
best_rf_auc = roc_auc_score(y_test, y_pred)

print(f"Best random-forest accuracy: {best_rf_acc:.3f}, f1-score: {best_rf_f1:.3f}, AUC: {best_rf_auc:.2f}, with {grid_res.best_params_['n_estimators']} estimators")
```

```python
pd.Series(y_pred).hist()
```

### Random-Forest Interpretation

In our best chosen model from above, it seems like RF performs decently with 74% accuracy and f1-score of 0.751, with ROC AUC 0.74

Previously I've tried to train the same model without re-balancing the data and the performances were much much worse, it seems like the re-balancing was necessary in order to get a decent performance here.

In addition by looking at the distribution of predicted probabilities we can see our model tends to classify more positive labels, this is also visible in the FP to TP ratio when looking at the confusion matrix.

## Gradient Boosting (LightGBM)

```python
feature_names = list(X.columns.values)
lgb_train = lgb.Dataset(X_train, y_train, feature_name=feature_names)
lgb_eval = lgb.Dataset(X_eval, y_eval, feature_name=feature_names)
```

```python
gbm = lgb.train(
    {
        "max_depth": 2,
        "is_unbalance": True,
    }, lgb_train, num_boost_round=500, valid_sets=lgb_eval, callbacks=[lgb.early_stopping(stopping_rounds=5)] )
```

```python
y_pred = gbm.predict(X_test)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve");
```

```python
pd.Series(y_pred).hist()
```

```python
print(confusion_matrix(y_test, y_pred>0.5))
best_gbm_acc = accuracy_score(y_test, y_pred>0.5)
best_gbm_auc = roc_auc_score(y_test, y_pred>0.5)
best_gbm_f1 = f1_score(y_test, y_pred>0.5)
```

```python
print(f"Best gradient boosting accuracy: {best_gbm_acc:.3f}, f1-score: {best_gbm_f1:.3f}, AUC: {best_gbm_auc:.2f}, with {gbm.best_iteration} boosting iterations")
```

### Gradient Boosting Interpretation
It looks like although we're getting some slightly better metrics with gradient boosting, the accuracy went up to 75% with f1-score of 0.763 and ROC AUC of 0.75. These are very similar to the performance of the previous model, but with a slightly lower FP and FN misclassifications.
Examining the predicted probability distributions here shows the same tendency of classifying more samples as positive labels.

We can examine some of the top features by importance (number of splits):

```python
# Plot feature importance

lgb.plot_importance(gbm, importance_type="gain", figsize=(7,6), title="LightGBM Feature Importance (Gain)")
plt.show()

lgb.plot_importance(gbm, importance_type="split", figsize=(7,6), title="LightGBM Feature Importance (Split)")
plt.show()

```

It looks like the most contributing features are BMI, GenHlth, Age and HighBP

This makes sense based on the earlier correlation we examined between the target variables and the predictors

# Conclusions

This project aimed to predict diabetes risk using the CDC Diabetes Health Indicators dataset. Through a series of preprocessing steps and model evaluations, both Random Forest and Gradient Boosting (LightGBM) models were implemented to classify participants. Gradient Boosting demonstrated slightly better performance, with an accuracy of 75%, F1 score of 0.763, and ROC AUC of 0.75, compared to Random Forestâ€™s accuracy of 74%, F1 score of 0.751, and AUC of 0.74.
What didn't work in the first attempts was training the same models on the imbalanced dataset, this resulted in very poor performances on the testing set.
Re-balancing the dataset was critical to improving model performance, particularly due to the class imbalance (14% diabetic/pre-diabetic).
Key predictors such as BMI, general health, and age were identified as top features for diabetes classification.
These results suggest that ensemble models, particularly gradient boosting, are effective in early prediction of diabetes risk and can be valuable in supporting healthcare decision-making for diabetes prevention.

Future work might include examination and comparison of additional type of models such as neural networks, and could incorporate additional data sources as well.
